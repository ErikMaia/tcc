```python

import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../assets/mes_data.csv")
df.columns
```

```python
len(df.columns)
```

```python
df["Informações do Passivo"]
```

```python
df = df.drop(columns=["Informações do Ativo","Informações do Passivo","Unnamed: 0"])
df
```

```python
df[df.isna()] = 0
df
```

```python
df = df.replace({'\.': '', '%': '', ',': '.'}, regex=True)
df['Dividend Yield do Mês de Referência³ (%)'] = pd.to_numeric(df['Dividend Yield do Mês de Referência³ (%)'])
df['Dividend Yield do Mês de Referência³ (%)']
```

```python
df['Dividend Yield do Mês de Referência³ (%)'].describe()
```

```python
sns.histplot(data=df.loc[(df['Dividend Yield do Mês de Referência³ (%)']>= -5) & (df['Dividend Yield do Mês de Referência³ (%)'] < 5)],x='Dividend Yield do Mês de Referência³ (%)',kde=True)
```

```python
sns.boxplot(data=df,x='Dividend Yield do Mês de Referência³ (%)')
```

```python
df.loc[df['Dividend Yield do Mês de Referência³ (%)'] <= 0 ]
```

```python
df.corr(method='pearson', min_periods=1, numeric_only=True)
```
